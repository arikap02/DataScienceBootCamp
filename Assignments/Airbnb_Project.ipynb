```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
'''Project Air BNB

1) Calculate How many properties in each neighbourhood group

2) Which Neighbour Hood Group Has highest prices use bar chart to show this for all neighbourhood groups

3) Which Neighbourhood group has lowest prices use bar chart to show this for all neighbourhood groups

4) Highest and lowest price airbnb in each Neighbour Hood Group

5) Which Host has highest count of properties

6) Which neighbourhood is most expensive and cheapest top 10 barchart

7) Count different type of room_type and do a pie chart showing % of each room type in overall dataset

8) scatter plot between longitutude and latitude for different neighbourhood groups

9) scatter plot between longitutude and latitude for different room_types

10) room_type vs price bar chart for the mean and highest and lowest

11) Do a similar plot to this'''
```

```python

#1) Calculate how many properties in each neighbourhood group

# Load the Airbnb data
df = pd.read_csv('AB_NYC_2019.csv')


# Calculate the number of properties in each neighbourhood group
n_properties_per_neighborhood_group = df['neighbourhood_group'].value_counts()

# Print the results
print(n_properties_per_neighborhood_group)
```

```python
#2) Which Neighbour Hood Group Has highest prices use bar chart to show this for all neighbourhood groups

# Calculate the average price per night for each neighbourhood group
average_price_per_night_per_neighborhood_group = df.groupby('neighbourhood_group')['price'].mean()

# Sort the neighbourhood groups by average price per night
average_price_per_night_per_neighborhood_group = average_price_per_night_per_neighborhood_group.sort_values(ascending=False)

# Get the neighbourhood group names
neighborhood_group_names = average_price_per_night_per_neighborhood_group.index.to_list()

# Get the average prices per night
average_prices_per_night = average_price_per_night_per_neighborhood_group.to_numpy()

# Create a bar chart of the average prices per night
plt.bar(neighborhood_group_names, average_prices_per_night)
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price per Night (USD)')
plt.title('Average Airbnb Prices in New York City by Neighbourhood Group')
plt.show()
```

```python

#3) Which Neighbourhood group has lowest prices use bar chart to show this for all neighbourhood groups

average_price_per_night_per_neighborhood_group = df.groupby('neighbourhood_group')['price'].mean()

# Sort the neighbourhood groups by average price per night
average_price_per_night_per_neighborhood_group = average_price_per_night_per_neighborhood_group.sort_values()

# Get the neighbourhood group names
neighborhood_group_names = average_price_per_night_per_neighborhood_group.index.to_list()

# Get the average prices per night
average_prices_per_night = average_price_per_night_per_neighborhood_group.to_numpy()

# Create a bar chart of the average prices per night
plt.bar(neighborhood_group_names, average_prices_per_night)
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price per Night (USD)')
plt.title('Average Airbnb Prices in New York City by Neighbourhood Group')
plt.show()
```

```python
#4) Highest and lowest price airbnb in each Neighbour Hood Group

# Calculate the highest and lowest price per night for each neighbourhood group
highest_price_per_night_per_neighborhood_group = df.groupby('neighbourhood_group')['price'].max()
lowest_price_per_night_per_neighborhood_group = df.groupby('neighbourhood_group')['price'].min()

# Get the neighbourhood group names
neighborhood_group_names = highest_price_per_night_per_neighborhood_group.index.to_list()

# Get the highest and lowest prices per night
highest_prices_per_night = highest_price_per_night_per_neighborhood_group.to_numpy()
lowest_prices_per_night = lowest_price_per_night_per_neighborhood_group.to_numpy()

# Print the results
print('Highest and lowest price Airbnb per neighbourhood group:')
for i in range(len(neighborhood_group_names)):
    print(f'{neighborhood_group_names[i]}: {highest_prices_per_night[i]} - {lowest_prices_per_night[i]}')
```

```python
#6) Which neighbourhood is most expensive and cheapest top 10 barchart

# Calculate the average price per night for each neighbourhood
average_price_per_night_per_neighborhood = df.groupby('neighbourhood')['price'].mean()

# Sort the neighbourhoods by average price per night
average_price_per_night_per_neighborhood = average_price_per_night_per_neighborhood.sort_values(ascending=False)

# Get the top 10 most expensive neighbourhoods
top_10_expensive_neighborhoods = average_price_per_night_per_neighborhood.index[:10]

# Get the top 10 cheapest neighbourhoods
top_10_cheapest_neighborhoods = average_price_per_night_per_neighborhood.index[-10:]

# Get the average prices per night for the top 10 neighbourhoods
top_10_expensive_average_prices_per_night = average_price_per_night_per_neighborhood[top_10_expensive_neighborhoods]
top_10_cheapest_average_prices_per_night = average_price_per_night_per_neighborhood[top_10_cheapest_neighborhoods]

# Create a bar chart of the average prices per night for the top 10 neighbourhoods
plt.bar(top_10_expensive_neighborhoods, top_10_expensive_average_prices_per_night)
plt.xlabel('Neighborhood')
plt.ylabel('Average Price per Night (USD)')
plt.title('Top 10 Most Expensive Airbnb Neighborhoods in New York City')
plt.show()

plt.bar(top_10_cheapest_neighborhoods, top_10_cheapest_average_prices_per_night)
plt.xlabel('Neighborhood')
plt.ylabel('Average Price per Night (USD)')
plt.title('Top 10 Cheapest Airbnb Neighborhoods in New York City')
plt.show()
```

```python
#7) Count different type of room_type and do a pie chart showing % of each room type in overall dataset

# Count the number of properties for each room type
room_type_counts = df['room_type'].value_counts()

# Calculate the percentage of each room type in the overall dataset
room_type_percentages = room_type_counts / df.shape[0] * 100

# Get the room type names
room_type_names = room_type_counts.index.to_list()

# Get the room type percentages
room_type_percentages = room_type_percentages.to_numpy()

# Create a pie chart of the room type percentages
plt.pie(room_type_percentages, labels=room_type_names, autopct="%1.1f%%")
plt.title("Percentage of Airbnbs by Room Type in New York City")
plt.show()
```

```python
#8) scatter plot between longitutude and latitude for different neighbourhood groups

# Create a scatter plot of latitude and longitude for each neighborhood group
plt.figure(figsize=(10, 5))
for neighborhood_group in df['neighbourhood_group'].unique():
    df_filtered = df[df['neighbourhood_group'] == neighborhood_group]
    plt.scatter(df_filtered['longitude'], df_filtered['latitude'], label=neighborhood_group)

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot of Airbnbs in New York City by Neighborhood Group')
plt.legend()
plt.show()
```

```python
#9) scatter plot between longitutude and latitude for different room_types

# Create a scatter plot of latitude and longitude for each room type
plt.figure(figsize=(10, 5))
for room_type in df['room_type'].unique():
    df_filtered = df[df['room_type'] == room_type]
    plt.scatter(df_filtered['longitude'], df_filtered['latitude'], label=room_type)

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot of Airbnbs in New York City by Room Type')
plt.legend()
plt.show()
```

```python
#10) room_type vs price bar chart for the mean and highest and lowest

# Calculate the mean, highest, and lowest price per night for each room type
room_type_mean_prices = df.groupby('room_type')['price'].mean()
room_type_highest_prices = df.groupby('room_type')['price'].max()
room_type_lowest_prices = df.groupby('room_type')['price'].min()

# Get the room type names
room_type_names = room_type_mean_prices.index.to_list()

# Create a bar chart of the mean, highest, and lowest price per night for each room type
plt.figure(figsize=(10, 5))

plt.bar(room_type_names, room_type_mean_prices, label='Mean Price')
plt.bar(room_type_names, room_type_highest_prices, label='Highest Price', color='red')
plt.bar(room_type_names, room_type_lowest_prices, label='Lowest Price', color='green')

plt.xlabel('Room Type')
plt.ylabel('Price per Night (USD)')
plt.title('Room Type vs Price in New York City')
plt.legend()
plt.show()
```
