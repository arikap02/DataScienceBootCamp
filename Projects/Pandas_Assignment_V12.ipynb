```python
#b. Sort the data by x1 in ascending order. What are the x1, x2, and x3 values of the first observation after the data are sorted?
df1 = df.sort_values(by=['x1'],ascending=True)
print(df1.iloc[0,:])
```

```python
#c. Sort the data by x1 in descending order. What are the x1, x2, and x3 values of the first observation after the data are sorted?
df2 = df.sort_values(by=['x1'],ascending=False)
print(df2.iloc[0,:])
```
